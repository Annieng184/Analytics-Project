```python
def initialize_driver():
    global DRIVER
    if DRIVER is None:
        print("Initiating driver...")
        chrome_options = webdriver.ChromeOptions()
        chrome_options.add_argument("-headless")
        chrome_options.add_argument("-no-sandbox")
        DRIVER = webdriver.Chrome(
            "chromedriver", options=chrome_options
        )  # Create the new chrome browser with specific options
        print("Finished!")
```

```python
def close_driver():
    global DRIVER
    if not DRIVER is None:
        DRIVER.quit()

    DRIVER = None
```

```python
def get_link_title_description(news_element):
  figure_element = news_element.find_element(By.CLASS_NAME, "figure")

  # Get the job title and article link
  figcaption_element = figure_element.find_element(By.CLASS_NAME, "figcaption")
  a_element = figcaption_element.find_element(By.TAG_NAME, "a")
  article_link = a_element.get_attribute("href")
  title = a_element.get_attribute("title")
  
  # Get the link to company and company name
  caption_element = figure_element.find_element(By.CLASS_NAME, "caption")
  a_caption_element = caption_element.find_element(By.TAG_NAME, "a")
  company_link = a_caption_element.get_attribute("href")
  company = a_caption_element.get_attribute("title")

  # Get the salary
  salary_element = figure_element.find_element(By.CLASS_NAME, "salary")
  p_element = salary_element.find_element(By.TAG_NAME, "p")
  salary = p_element.text

  # Get the location
  location_element = figure_element.find_element(By.CLASS_NAME, "location")
  ul_element = location_element.find_element(By.TAG_NAME, "ul")
  li_element = ul_element.find_element(By.TAG_NAME, "li")
  location = li_element.text

  # Get the time
  time_element = figure_element.find_element(By.CLASS_NAME, "time")
  date_element = time_element.find_element(By.TAG_NAME, "time")
  date = date_element.text

  return title, article_link, company, company_link, salary, location, date
```

```python
def scrape_careerbuilder(DRIVER):

    results = []
    all_news_elements = DRIVER.find_elements(By.CLASS_NAME, "job-item")

    # Optional to keep track
    n_skip = 0
    n_success = 0

    for news_element in all_news_elements:
        try:
            title, article_link, company, company_link, salary, location, date = get_link_title_description(news_element)  # must have 7
            n_success += 1
        except Exception:
            print("Cannot get this article. Skip")
            n_skip += 1
            continue

        
        results.append([title, article_link, company, company_link, salary, location, date])

    print(f"Number of scraped articles: {n_success}")
    print(f"Number of skipped articles: {n_skip}")

    return results
```

```python
results = scrape_careerbuilder(DRIVER)
```

```python
def scrape_careerbuilder_page(driver, page_number):
  url = f"https://careerbuilder.vn/viec-lam/tat-ca-viec-lam-trang-{page_number}-vi.html"
  driver.get(url)
  return scrape_careerbuilder(driver)

results = []
for page_number in range(1, 600):
  results += scrape_careerbuilder_page(DRIVER, page_number)
```
